## 2. The West Coast Heat Wave

In 2016 [Dr. Chelle Gentemann](https://cgentemann.github.io) and collaborators published a [paper](https://agupubs.onlinelibrary.wiley.com/doi/10.1002/2016GL071039) on the heat wave in the ocean off the California coast

The analysis was then performed using Matlab to process scientific data. To make Figure 1, here are the following steps:
- Download 4 TB of data from NASA PO.DAAC data archive via FTP
- Go through each day of data and subset to the West Coast Region to reduce size and save each subsetted day
- Go through 2002-2012 and create a daily climatology and save all 365 days of the climatology
- Go through each day of data and calculate the anomaly and save each day's anomaly

This whole process took about 1-2 month. 
Below we will do this using MUR SST data on AWS Open Data Program in a few minutes using Python.

### Question 4 

Reproduce Figure 1 in the paper. When doing that, try to slice data using `xarray` in the most clear and reproducible way (for example, don't use indices to access data objects like we did in the example above).

```python
#calculate SST anomolies
anomolies = ds.analysed_sst - ds.analysed_sst.mean("time")

#plot subplots for each month
nmonths=12
nyears=3
fig, axs = plt.subplots(nrows=nmonths, ncols=nyears, sharex=True, sharey=True, subplot_kw={'projection': ccrs.PlateCarree()}, figsize=(9, 20))
for i in range(nmonths):
    for j in range(nyears):
        
        # find the coordinates
        timestr=f"20{14+j}-{i+1:02d}"
        month_name = pd.to_datetime(timestr, format="%Y-%m").month_name()[:3]
        
        #plot SST anomoly by lat and lon
        anoma_map = anomolies.sel(time=timestr).mean("time").plot(ax=axs[i, j], transform=crs, cmap='RdBu_r', 
                                                                          add_colorbar=False) 
        axs[i, j].add_feature(cfeature.LAND, color="grey")
        axs[i, j].add_feature(cfeature.STATES.with_scale("10m"))
        axs[i, j].coastlines("10m", color="k")
        
        # labeling each graph
        axs[i, j].text(-124.5, 45, month_name, fontsize=15, color="white", fontweight="semibold")
        axs[i, j].set(xlabel=None, ylabel=None)
        
        # labels / ticks based on subplot position
        if i == 0:
            axs[i, j].set_title(f"{timestr[:4]}")
        if j == 0:
            axs[i, j].set_yticks([35, 40, 45])
        if i == 11:
            axs[i, j].set_xticks([-130, -120])

#spacing of subplots
fig.subplots_adjust(left=0, hspace=.215, wspace=-.77)

# add colorbar
ax = fig.add_axes((.235, .08, .42, .01))
cbar = plt.colorbar(anoma_map, cax=ax, orientation='horizontal')
cbar.set_label('SST(°C)', loc="center")
cbar.set_ticks([-2, 0, 2])
cbar.set_ticklabels([-2, 0, 2])

fig.show();
```
